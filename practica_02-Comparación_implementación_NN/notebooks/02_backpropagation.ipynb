Paso 1: Entender la estructura básica
- Una red neuronal tiene:

Entrada → Capa Oculta → Salida  
(784)-------(128)--------(10)

¿Por qué estas capas?

- Entrada (784): Un píxel = una neurona (28×28 = 784)
- Capa Oculta (128): Aprende patrones intermedios (curvas, bordes)
- Salida (10): Una neurona por dígito (0-9)


Paso 2: Lo que necesitas implementar
- Inicializar pesos (matrices W1, W2) y biases (b1, b2)
- Forward pass: calcular la salida
- Función de activación: sigmoid o ReLU
- Calcular error (loss)
- Backward pass: calcular gradientes
- Actualizar pesos: gradient descent
